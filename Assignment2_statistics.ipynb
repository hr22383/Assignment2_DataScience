```python
dir_path =  r"//content/drive/MyDrive/CE888/EyeT"
```

```python

import glob

csv_files = glob.glob(os.path.join(dir_path, "*.csv"))

df = pd.DataFrame()
  
# loop over the list of csv files
for file in csv_files:  
  # read the csv file
  temp_df = pd.read_csv(file, header=10, nrows=3) # 3 sample per file; total 502 files; total of 1506 samples
  df = df.append(temp_df, ignore_index=True)

df.head()
```

```python
# Displaying descriptive statistics 
df.describe()
```

```python
# Importing Library for Co Relation Matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix with given Dataset
corr_matrix = df.corr()
```

```python
# Printing Co Relation Matrix
corr_matrix
```

```python
# Create a heatmap using seaborn for Co Relation Matrix
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)
# Show the plot
plt.show()
```
